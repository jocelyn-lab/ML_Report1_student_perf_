```python
y = data['G3'] # Target variable y 
X = data.drop(columns=['G3', 'G1', 'G2']) # Features X (all columns except 'G3', 'G1', and 'G2')
X = pd.get_dummies(X, drop_first=True) # Convert categorical variables to dummy variables
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42) # Split the data into training and testing sets (80% train, 20% test)
model = LinearRegression() # Create a linear regression model
model.fit(X_train, y_train) # Fit the model to the training data
y_pred = model.predict(X_test) # Predict the target variable for the test set
```
